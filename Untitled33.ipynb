```python
PATCH = 64

def make_patches(img):
    patches = []
    H, W, _ = img.shape
    for r in range(0, H, PATCH):
        for c in range(0, W, PATCH):
            if r+PATCH <= H and c+PATCH <= W:
                patches.append(img[r:r+PATCH, c:c+PATCH])
    return patches
```

```python
import numpy as np
from PIL import Image
import os

IMG_ROOT = "/content/extracted/images/Onera Satellite Change Detection dataset - Images"
LABEL_ROOT = "/content/extracted/train_labels/Onera Satellite Change Detection dataset - Train Labels"

def load_city(city):
    img1 = np.array(Image.open(f"{IMG_ROOT}/{city}/pair/img1.png"))
    img2 = np.array(Image.open(f"{IMG_ROOT}/{city}/pair/img2.png"))
    mask = np.array(Image.open(f"{LABEL_ROOT}/{city}/cm/cm.png").convert("L"))

    # convert to binary 0/1
    mask = (mask > 128).astype(np.uint8)

    return img1, img2, mask
```

```python
def extract_features(img1, img2):
    img1 = img1.astype(np.float32)
    img2 = img2.astype(np.float32)

    diff = img2 - img1
    ratio = img2 / (img1 + 1e-5)

    # fake NDVI for RGB
    ndvi1 = (img1[:,:,1] - img1[:,:,0]) / (img1[:,:,1] + img1[:,:,0] + 1e-5)
    ndvi2 = (img2[:,:,1] - img2[:,:,0]) / (img2[:,:,1] + img2[:,:,0] + 1e-5)
    ndvi_diff = (ndvi2 - ndvi1)[...,None]

    # flatten all features
    feat = np.concatenate([diff, ratio, ndvi_diff], axis=-1)
    H, W, C = feat.shape
    return feat.reshape(-1, C)
```

```python
TRAIN_CITIES = ["mumbai","beirut","bordeaux","nantes","paris"]

X_all = []
Y_all = []

for city in TRAIN_CITIES:
    img1, img2, mask = load_city(city)
    feat = extract_features(img1, img2)
    lab  = mask.reshape(-1)

    X_all.append(feat)
    Y_all.append(lab)

X_all = np.concatenate(X_all)
Y_all = np.concatenate(Y_all)

print(X_all.shape, Y_all.shape)
```

```python
import numpy as np
from sklearn.model_selection import train_test_split

# -----------------------------
# 1) FIX LABELS (convert >0 → 1)
# -----------------------------
Y_all = (Y_all > 0).astype(np.uint8)

print("Unique labels in whole dataset:", np.unique(Y_all, return_counts=True))

# -----------------------------
# 2) BALANCE THE DATA
# -----------------------------
idx_0 = np.where(Y_all == 0)[0]
idx_1 = np.where(Y_all == 1)[0]

take_1 = min(len(idx_1), 50000)   # max 50k positive
take_0 = take_1 * 3               # 3× negative

idx_1_sample = np.random.choice(idx_1, take_1, replace=False)
idx_0_sample = np.random.choice(idx_0, take_0, replace=False)

idx_final = np.concatenate([idx_0_sample, idx_1_sample])
np.random.shuffle(idx_final)

X_sample = X_all[idx_final]
Y_sample = Y_all[idx_final]

print("Balanced shape:", X_sample.shape, Y_sample.shape)

# -----------------------------
# 3) TRAIN–TEST SPLIT
# -----------------------------
X_train, X_val, y_train, y_val = train_test_split(
    X_sample, Y_sample, test_size=0.2, random_state=42
)

print("Train:", X_train.shape, y_train.shape)
print("Val:",   X_val.shape,   y_val.shape)
```

```python
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
import lightgbm as lgb

# RF
rf = RandomForestClassifier(n_estimators=150)
rf.fit(X_train, y_train)

# XGB
xgb = XGBClassifier(
    n_estimators=200,
    max_depth=6,
    eval_metric="logloss"
)
xgb.fit(X_train, y_train)

# LGBM
lgbm = lgb.LGBMClassifier(n_estimators=200)
lgbm.fit(X_train, y_train)

print("RF trained")
print("XGB trained")
print("LGBM trained")
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, jaccard_score

p1 = rf.predict_proba(X_val)[:,1]
p2 = xgb.predict_proba(X_val)[:,1]
p3 = lgbm.predict_proba(X_val)[:,1]

# Ensemble voting
p_final = (p1 + p2 + p3) / 3
pred_bin = (p_final > 0.5).astype(int)

print("Accuracy :", accuracy_score(y_val, pred_bin))
print("Precision:", precision_score(y_val, pred_bin))
print("Recall   :", recall_score(y_val, pred_bin))
print("F1 Score :", f1_score(y_val, pred_bin))
print("IoU      :", jaccard_score(y_val, pred_bin))
```
