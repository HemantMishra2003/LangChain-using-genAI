```python
#  important libraries

import rasterio # use for reading settelite images.
import numpy as np # helful for making array
import os          # use for making foder

city = "mumbai"  # here we are defining city to load



# ROOT paths,here we are defining our settlelite image and labelled images
img_root = "/content/extracted/images/Onera Satellite Change Detection dataset - Images"
label_root = "/content/extracted/train_labels/Onera Satellite Change Detection dataset - Train Labels"

# defining our main files , whom we have to work.
t1_path = f"{img_root}/{city}/imgs_1_rect"
t2_path = f"{img_root}/{city}/imgs_2_rect"
label_path = f"{label_root}/{city}/cm/{city}-cm.tif" # this files told where
# the change has occured  and where not.

print("T1 folder:", t1_path)
print("T2 folder:", t2_path)
print("Label path:", label_path)

# List band files
t1_files = sorted(os.listdir(t1_path))
t2_files = sorted(os.listdir(t2_path))

print("Bands in t1:", t1_files)

# Loading here t1 before images
t1 = []
for f in t1_files:
    with rasterio.open(f"{t1_path}/{f}") as src:
        t1.append(src.read(1))
t1 = np.stack(t1, axis=-1)

# Load t2 after images
t2 = []
for f in t2_files:
    with rasterio.open(f"{t2_path}/{f}") as src:
        t2.append(src.read(1))
t2 = np.stack(t2, axis=-1)

# Loading  label
with rasterio.open(label_path) as src:
    label = src.read(1)

print("T1:", t1.shape)
print("T2:", t2.shape)
print("LABEL:", label.shape)
```

```python
# Normalize function
def norm(x):
    x = x.astype('float32')
    x = x - x.min()
    x = x / (x.max() + 1e-6)
    return x

# Normalize all 13 bands for T1 and T2
t1_norm = norm(t1)
t2_norm = norm(t2)

# Combine both into 26 channels
combined = np.concatenate([t1_norm, t2_norm], axis=-1)

print("Combined shape:", combined.shape)
print("Label shape:", label.shape)
```

```python
PATCH = 128
X = []
Y = []

h, w, _ = combined.shape

for i in range(0, h-PATCH, PATCH):
    for j in range(0, w-PATCH, PATCH):

        x_patch = combined[i:i+PATCH, j:j+PATCH, :]
        y_patch = label[i:i+PATCH, j:j+PATCH]

        X.append(x_patch)
        Y.append(y_patch)

X = np.array(X)
Y = np.array(Y)

print("X shape:", X.shape)
print("Y shape:", Y.shape)
```

###  train test - Split

```python
from sklearn.model_selection import train_test_split

X_train, X_val, Y_train, Y_val = train_test_split(
    X, Y, test_size=0.2, random_state=42
)

print(X_train.shape, X_val.shape)
```

```python
import tensorflow as tf
from tensorflow.keras import layers, models

def conv_block(x, filters):
    x = layers.Conv2D(filters, 3, padding="same")(x)
    x = layers.BatchNormalization()(x)
    x = layers.ReLU()(x)

    x = layers.Conv2D(filters, 3, padding="same")(x)
    x = layers.BatchNormalization()(x)
    x = layers.ReLU()(x)
    return x

def encoder_block(x, filters):
    f = conv_block(x, filters)
    p = layers.MaxPooling2D((2,2))(f)
    return f, p

def decoder_block(x, skip, filters):
    x = layers.UpSampling2D((2,2))(x)
    x = layers.Concatenate()([x, skip])
    x = conv_block(x, filters)
    return x

def build_unet(input_shape=(128,128,26)):
    inputs = layers.Input(input_shape)

    # Encoder
    f1, p1 = encoder_block(inputs, 32)
    f2, p2 = encoder_block(p1, 64)
    f3, p3 = encoder_block(p2, 128)
    f4, p4 = encoder_block(p3, 256)

    # Bottleneck
    bn = conv_block(p4, 512)

    # Decoder
    d1 = decoder_block(bn, f4, 256)
    d2 = decoder_block(d1, f3, 128)
    d3 = decoder_block(d2, f2, 64)
    d4 = decoder_block(d3, f1, 32)

    # Output layer
    outputs = layers.Conv2D(1, 1, activation="sigmoid")(d4)

    model = models.Model(inputs, outputs)
    return model

unet = build_unet()
unet.summary()
```

```python
def dice_loss(y_true, y_pred):
    smooth = 1
    y_true_f = tf.reshape(y_true, [-1])
    y_pred_f = tf.reshape(y_pred, [-1])
    intersection = tf.reduce_sum(y_true_f * y_pred_f)
    return 1 - ((2.*intersection + smooth) /
                (tf.reduce_sum(y_true_f) + tf.reduce_sum(y_pred_f) + smooth))

unet.compile(
    optimizer = tf.keras.optimizers.Adam(1e-4),
    loss = dice_loss,
    metrics = ["accuracy"]
)
```

#### Model Training

```python
import os
import numpy as np
import rasterio
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
import lightgbm as lgb
import matplotlib.pyplot as plt

IMG_ROOT = "/content/extracted/images/Onera Satellite Change Detection dataset - Images"
LABEL_ROOT = "/content/extracted/train_labels/Onera Satellite Change Detection dataset - Train Labels"
```

```python
TRAIN_CITIES = [
    "mumbai","beirut","bordeaux","bercy","rennes","saclay_e",
    "abudhabi","cupertino","nantes","paris","beihai","hongkong",
    "aguasclaras","pisa"
]

TEST_CITIES = [
    "dubai","lasvegas","rio","montpellier","norcia",
    "valencia","saclay_w","milano","chongqing","brasilia"
]
```

```python
def load_city(city, is_train=True):
    t1_path = f"{IMG_ROOT}/{city}/imgs_1_rect"
    t2_path = f"{IMG_ROOT}/{city}/imgs_2_rect"

    # Load T1
    t1 = np.stack([
        rasterio.open(f"{t1_path}/{b}").read(1)
        for b in sorted(os.listdir(t1_path))
    ], axis=-1)

    # Load T2
    t2 = np.stack([
        rasterio.open(f"{t2_path}/{b}").read(1)
        for b in sorted(os.listdir(t2_path))
    ], axis=-1)

    if not is_train:
        return t1, t2, None

    # Load label CORRECTLY
    label_path = f"{LABEL_ROOT}/{city}/cm/{city}-cm.tif"
    with rasterio.open(label_path) as src:
        lab = src.read(1)

    # ⭐ FIX 1: Convert {-1, 0, 1} → {0, 1}
    lab = np.where(lab > 0, 1, 0).astype(np.uint8)

    return t1, t2, lab
```

```python
t1, t2, lab = load_city("mumbai", True)
print(np.unique(lab))
```

```python
def extract_features(t1, t2):
    t1 = t1.astype(float)
    t2 = t2.astype(float)

    diff = t2 - t1
    ratio = t2 / (t1 + 1e-6)

    ndvi_t1 = (t1[:,:,7] - t1[:,:,3]) / (t1[:,:,7] + t1[:,:,3] + 1e-6)
    ndvi_t2 = (t2[:,:,7] - t2[:,:,3]) / (t2[:,:,7] + t2[:,:,3] + 1e-6)

    ndbi_t1 = (t1[:,:,10] - t1[:,:,7]) / (t1[:,:,10] + t1[:,:,7] + 1e-6)
    ndbi_t2 = (t2[:,:,10] - t2[:,:,7]) / (t2[:,:,10] + t2[:,:,7] + 1e-6)

    ndwi_t1 = (t1[:,:,2] - t1[:,:,7]) / (t1[:,:,2] + t1[:,:,7] + 1e-6)
    ndwi_t2 = (t2[:,:,2] - t2[:,:,7]) / (t2[:,:,2] + t2[:,:,7] + 1e-6)

    full_feat = np.concatenate([
        diff,                     # 13 bands
        ratio,                    # 13 bands
        (ndvi_t2-ndvi_t1)[...,None],
        (ndbi_t2-ndbi_t1)[...,None],
        (ndwi_t2-ndwi_t1)[...,None],
    ], axis=-1)

    H, W, C = full_feat.shape
    return full_feat.reshape(-1, C)
```

```python
X_all = []
Y_all = []

for city in TRAIN_CITIES:
    print("Loading:", city)
    t1, t2, lab = load_city(city, True)

    feats = extract_features(t1, t2)
    labels = lab.reshape(-1)

    X_all.append(feats)
    Y_all.append(labels)

X_all = np.concatenate(X_all)
Y_all = np.concatenate(Y_all)

print("Final dataset:", X_all.shape, Y_all.shape)
```

```python
# BALANCED SAMPLING FIX

# Check counts
idx_0 = np.where(Y_all == 0)[0]
idx_1 = np.where(Y_all == 1)[0]

print("0 count:", len(idx_0))
print("1 count:", len(idx_1))

# Balanced sampling
N = 100000   # choose balanced pixels

take_0 = np.random.choice(idx_0, N//2, replace=False)
take_1 = np.random.choice(idx_1, N//2, replace=False)

balanced_idx = np.concatenate([take_0, take_1])
np.random.shuffle(balanced_idx)

# final sampled dataset
X_sample = X_all[balanced_idx]
Y_sample = Y_all[balanced_idx]

print("Balanced labels:", np.unique(Y_sample, return_counts=True))
```
